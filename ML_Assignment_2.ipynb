```python
#Importing packages

import pandas as pd
import numpy as np
from IPython.display import display
```

1. Reading CSV Dataset

```python
# Load Dataset URL
csv_url = 'https://drive.google.com/uc?id=1BRkN4QGpcYQjhPZmdUE-j73HKh4UjhWX'
```

```python
# Reading and displaying hosted(In google drive) dataset

df = pd.read_csv(csv_url)
display(df[:15])
```

#2. Handling missing values

```python
#Replacing "?" value with NaN inorder to handle the missing values

df.replace("?", np.nan, inplace=True)
print(df.isnull().sum())
```

```python
total_rows = df.shape[0] -1
total_columns = df.shape[1]

print(total_rows)
```

```python
rows_with_any_nan_value = df.isna().any(axis=1)
```

```python
#Displaying rows with missing values

display(df[rows_with_any_nan_value])
```

```python
# Handling missing values with appropriate methods

df["Workclass"].fillna(df["Workclass"].mode()[0], inplace=True)
df["Occupation"].fillna(df["Occupation"].mode()[0], inplace=True)

df["Native Country"].fillna("Unknown", inplace=True)
```

```python
display(df[rows_with_any_nan_value])
```

#3. Convert Categorical Data into numerical form

```python
# Convert categorical data into numerical form using one-hot encoding

df_encoded = pd.get_dummies(df, columns = ['Workclass', 'Marital Status', 'Occupation', 'Relationship', 'Race', 'Native Country']);

display(df_encoded)
```

```python
#Convert binary categorical data to 1 and 0

df_encoded['Sex'] = (df_encoded['Sex'] == 'Male').astype(int)
df_encoded['Income'] = (df_encoded['Income']=='>50K').astype(int)

#Converting all boolean values into 1's and 0's
df_encoded = df_encoded.applymap(lambda x : 1 if x is True else (0 if x is False else x))

# Drop all columns that is not in numerical form
for col in df_encoded.columns:
  if not pd.api.types.is_numeric_dtype(df_encoded[col]):
    df_encoded = df_encoded.drop(col, axis = 1)


display(df_encoded)
```

#4. Scale/Normalize numerical feature

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Scale some features using Min-Max Scaler

df_encoded['Age'] = scaler.fit_transform(df_encoded[['Age']])
df_encoded['Hours Per Week'] = scaler.fit_transform(df_encoded[['Hours Per Week']])
df_encoded['Capital Gain'] = scaler.fit_transform(df_encoded[['Capital Gain']])
df_encoded['Capital Loss'] = scaler.fit_transform(df_encoded[['Capital Loss']])

display(df_encoded)
```

#5. Detect and handle outliers

#6. Feature Engineering

```python
#Generate Correlation matrix and displaying all rows which are correlation value larger than 0.3

corr_matrix = df_encoded.corr()

corr_matrix = corr_matrix[(abs(corr_matrix)) > 0.3 ];
corr_matrix = corr_matrix[(abs(corr_matrix)) != 1.0 ];

corr_matrix = corr_matrix.dropna(how='all', axis = 0)
corr_matrix = corr_matrix.dropna(how='all', axis = 1)

display(corr_matrix)
```

#7. Split Data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Drop target Column
X = (df_encoded.drop('Income', axis=1)).to_numpy()
Y = (df_encoded['Income']).to_numpy()

# Scale feature columns using StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split train and test data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=42)

# Load model
model = LogisticRegression()

# Train logistic regression model
history = model.fit(x_train, y_train)
```

```python
# Evaluate model accuracy using test data

predictions = model.predict(x_test)

correct_predictions = np.sum(predictions == y_test)
total_predictions = len(predictions)
correct_prediction_precentage = (correct_predictions/total_predictions)*100
print(f'Accuracy precentage : {correct_prediction_precentage:.2f}%')
```

#7. Evaluate the result

```python
# Load other matrics for evaluate model's accuracy

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve, log_loss
```

```python
y_pred = np.array(predictions)
y_true = np.array(y_test)
```

```python
print("Accuracy : ", accuracy_score(y_true, y_pred))
```

```python
print("Precision : ", precision_score(y_true, y_pred))
```

```python
print("Recall : ", recall_score(y_true, y_pred))
```

```python
print("F1 Score : ", f1_score(y_true, y_pred))
```

```python
print("Confution matrix :\n ", confusion_matrix(y_true, y_pred))
```

#8. Train a model with neural network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input

# Neural network model creation

nn_model = Sequential([
    Input(shape= (x_train.shape[1],)),
    Dense(units=128, activation = 'relu'),
    Dropout(0.5),
    Dense(units=64, activation = 'relu'),
    Dropout(0.5),
    Dense(units = 32, activation='relu'),
    Dense(units=1, activation='sigmoid')
])
```

```python
# Compile NN model with defining optimizer and loss function

nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training NN model 

history = nn_model.fit(x_train, y_train, epochs=50, batch_size=64, validation_data=(x_test, y_test))
```

```python
# Evaluate NN model accuracy using test data

nn_prediction_probabilities = nn_model.predict(x_test)
```

```python
nn_prediction_probabilities = nn_model.predict(x_test)
nn_predictions = np.array((nn_prediction_probabilities > 0.5).astype(int))
nn_predictions = np.ravel(nn_predictions) # (13555,) format

nn_correct_predictions = np.sum(nn_predictions == y_test)
nn_total_predictions = len(nn_predictions)
nn_accuracy = (nn_correct_predictions/nn_total_predictions)*100

nn_y_pred = np.array(nn_predictions)
nn_y_true = np.array(y_test)


print(f'NN Accuracy : {nn_accuracy:.2f}%')

```

```python
print("Accuracy : ", accuracy_score(nn_y_true, nn_y_pred))
```

```python
print("Precision : ", precision_score(nn_y_true, nn_y_pred))
```

```python
print("Recall : ", recall_score(nn_y_true, nn_y_pred))
```

```python
print("F1 Score : ", f1_score(nn_y_true, nn_y_pred))
```

```python
print("Confution matrix :\n ", confusion_matrix(nn_y_true, nn_y_pred))
```
